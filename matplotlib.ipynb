```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import lineStyles

mpl.rcParams.update({
    'font.family':'SimHei', # or 'STSong' to type Chinese
    'figure.dpi': 100
})
```

# figure and axes.

```python
fig, ax = plt.subplots(figsize=(5, 2.7))             # Create a figure containing a single Axes.
fig.suptitle('画布')
plt.show()
fig.savefig('MyFigure.png', dpi=200)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
np.random.seed(19680801)
t = np.arange(200)
x = np.cumsum(np.random.randn(200))
y = np.cumsum(np.random.randn(200))
linesx = ax.plot(t, x, label='Random walk x', color="r", linewidth=2, linestyle=':')
linesy = ax.plot(t, y, label='Random walk y', color="k", linestyle="--")

ax.set_xlabel('Time [s]')
ax.set_ylabel('Distance [km]')
ax.set_title('Random walk example')
ax.legend()
plt.show()
```
